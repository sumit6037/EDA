```python
from tabulate import tabulate

def sep(df):
    cateogrical=[]
    continuous=[]
    for col in df.columns:
        if df[col].nunique()<100:
            cateogrical.append(col)
        else:
            continuous.append(col)
    return cateogrical,continuous
categorical,continous=sep(df)

table=[categorical,continous]
print(tabulate({"Categorical":categorical,
                "Continous":continous
                },headers=["Categorical","Continuous"]))
            
```


# $$\;\;\;\; Univariate \ Analysis \ of \ Categorical \ Variables$$

```python
def info_of_cat(col):
    print("Unique values in",col,"are: ",df[col].unique())
    print("Mode of", col,"is: ",df[col].mode()[0])
    print("No of missing values in", col,"is",df[col].isnull().sum())
    if df[col].isnull().sum()>0:
        print("\nThere are null values in the",col,"column")
```

```python
def info_of_num(col):
    print("Mean of the ",col," is: ",df[col].mean())
    print("Median of the ",col," is: ",df[col].median())
    print("Mode of the ",col," is: ",df[col].mode()[0])
    print("Standard deviation of the ",col," is: ",df[col].std())
    print("Missing values of ",col," are: ",df[col].isnull().sum())
```
